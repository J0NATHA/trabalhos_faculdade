# **FOLIUM**

## **Introduo**

```python
import folium
```

```python
latitude = -22.9007476
longitude = -43.5815068
coordenadas = [latitude, longitude]
uezo = coordenadas
```

```python
mapa = folium.Map(location=uezo, star_zoom=9)
```

```python
folium.Marker(uezo, popup='UEZO').add_to(mapa)
```

```python
mapa # mostra o mapa
```

## **Marcar:**
*   O estdio do seu time de futebol;
*   O Cristo redentor;
*   As Pirmides do egito

```python
maracan = [-22.9130958, -43.2313389]
cristo_redentor = [-22.9519139,-43.2115823]
pirmides_egito = [29.9773008,31.1299206]
```

```python
folium.Marker(maracan, popup='Maracan').add_to(mapa)
folium.Marker(cristo_redentor, popup='Cristo Redentor').add_to(mapa)
folium.Marker(pirmides_egito, popup='Pirmides do Egito').add_to(mapa)
```

```python
folium.GeoJson('Limite_de_Bairros.geojson').add_to(mapa)
```

```python
mapa.save('mapa.html') # salva o mapa em um arquivo html
mapa
```
